```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("epicurve.xlsx", usecols=["Onset Date", "Case Count"])
```

```python
df['Onset Date'] = pd.to_datetime(df['Onset Date'])
df = df.set_index(df['Onset Date'])
```

```python
df['pandas_SMA_7'] = df.iloc[:,1].rolling(window=7).mean()
```

```python
x = df['Onset Date']
```

```python
y = df['pandas_SMA_7']
```

```python
plt.figure(figsize=[15,10])

x = x
y1 = y
y2 = df['Case Count']

plt.plot(x,y1, label="7 day moving average")
plt.plot(x,y2, label = "Daily case count")
plt.xlabel("Date")
plt.ylabel("Case Count")
plt.legend(loc='lower right')
```
